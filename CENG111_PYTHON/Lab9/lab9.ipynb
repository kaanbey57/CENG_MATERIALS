# Example1 Factorial

Lets define factorial() function that we studied in the last lab with iterations.

````python
def factorial(n):
    result=1
    for i in range(2,n+1):
        result*=i
    return result
print(factorial(4))
````

# Example2 (Print Triangle)

This is a classic example. Lets define a function triangle(h) where we print a triangle of height h with "*" symbol

```python
>>>triangle(5)
         *         
       * * *       
     * * * * *     
   * * * * * * *   
 * * * * * * * * *
```

````python
def triangle(n):
    for i in range(1,n+1):
        print(" "*(n-i)+"*"*(2*i-1))

def triangle2(h):
    for i in range(h):
        for j in range(h-i-1):
            print(' ', end='')
        for j in range(2*i+1):
            print('*', end='')
        for j in range(h-i-1):
            print(' ', end='')
        print('')    
        
triangle(5)
triangle2(5)
````

# Example3

Define a function findNumber that returns the index of the first number in list L. If not found, return -1


````python
def findNumber(lst):
    for i in range(len(lst)):
        if type(lst[i]) in (float,int):
            return i
            break
    return -1
    
print(findNumber(['a','b',1,'c','d']))
````

## Loop with else statement

We can use `else` in loops like this;

```python

while ...:
    ..
    ..
    
else:
    ...
    
    
or

for ...:
    ..
    ..
else:
    ...
```

The statements in the else part are executed when:

1. Test condition becomes false in while loop
2. list is exhausted in for loop

i.e. if loop ends with a **break** statement, **else** part is **not executed**.

Define a function andAll which takes a list of boolean expressions and return True if all elements of list are True. Note: assume all elements of L are valid booleans.

````python
def andAll(L):
    for el in L:
        if not el:
            result= False
            break
    else:
        result=True
    return result

print(andAll([2==2,3<5,7!=8]))
print(andAll([2==2,5>6, 'a'=='a']))
````

# Example6

Let's define a function called isperfect which takes an integer and returns True if;

- the number is perfect, false otherwise

- A number n is perfect if it is equal to the sum of its proper divisors

````python
def isPerfect(n):
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum+=i
    return sum==n
print(isPerfect(7))
print(isPerfect(6))
print(isPerfect(28))
````

# Example7 (Bubble Sort)

Let's define one of the simplest sorting algorithms, bubble sort. The logic is swapping adjacent elements if their order is wrong.

````python
exampleList=[6,5,4,3,2,10,1]

def bubble(lst):
    n=len(lst)
    for i in range(n):
        swapped=False
        for j in range(n-i-1):
            if lst[j]>lst[j+1]:
                lst[j], lst[j+1]=lst[j+1],lst[j]
                swapped=True
        if swapped == False:
            break
bubble(exampleList)
print(exampleList)
````
