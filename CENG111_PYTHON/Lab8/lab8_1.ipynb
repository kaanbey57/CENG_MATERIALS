## Different Ways to Work With a Script

In python, we can use scripts (`.py`) files for:

- Only defining `functions`,`classes` etc., and use them in another `.py` files by importing
- Only running expressions and calling functions
- Both of above

For example, if we have function defs, calls and some statements in a file called `example.py`

```python

def someFunc():
    ...
    ...

def someFunc2():
    ...
    ...

someFunc()

someFunc2()

x = 5

print(x)

```

and we import this code in another `.py` file called `callExample.py` like;

```python

import example

```

This is going to run `example.py` and it will execute `someFunc()` , `someFunc2()` and do the `print`. If we do not want to run anything but only import `someFunc()` and `someFunc2()` while **importing**, we have to use statements and function calls inside `if __name_ == "__main":` block like this in `example.py`

```python

def someFunc():
    ...
    ...

def someFunc2():
    ...
    ...

if __name__ = "__main__":
    someFunc()
    someFunc2()

    x = 5

    print(x)


```

If we now import this in `callExample.py`

```python
import example
```

We will only import the functions and do not run the code. The code only runs if we call it from terminal like `python3 example.py`

## Higher Order Functions (map, filter reduce)

Higher order functions take another **function** as an argument

## MAP

Map function takes a unary function (function that takes one argument) and a container type (list, tuple etc.), applies the given function to each element

## FILTER

Filter function takes a unary function that returns a boolean and a container type. It returns a new container by eliminating the original elements returning False for the given function

- Let's write a function named `keepNonNegatives` that takes a list of integers and returns a new list which has the non negative integers from the original list.

````python
def isNonNegative(x):
    return abs(x) == x
def keepNonNegatives(lst):
    return(list(filter(isNonNegative, lst)))
print(keepNonNegatives([1,2,3,0,-10,-5]))
````

## REDUCE

Reduce function takes a binary function, a container type and optionally an initial value. It reduces the given container to a single cumulative value by the application of the given function. It's a built-in function in Python2, but must be imported from functools in Python3

### Reduce Sum

Let's write a function named sumElements using reduce() to calculate the summation of a given list of integers

````python
from functools import reduce

def myAdd(a,b):
    print(str(a)+"+"+str(b)+"="+str(a+b))
    return a+b
def sumElements(lst):
    return reduce(myAdd, lst)

print(sumElements([1,2,3,4,5]))
````

## Lambda Expressions

We can write inline and anonymous functions without any names using lambda expressions. We can do this like this;

```python

lambda arguments : expression

```

Let's write an alternative solution for `sumElements`

**LAMBDA CALCULUS**

````python
def sumElements(lst):
    return reduce(lambda a,b : a+b, lst)

print(sumElements([1,2,3,4,5]))
````

## Filter Users

Let's assume that we have a database of users where each user is a tuple of;

- Id
- Username
- Age

Let's write a function called `filterUsers` where we return a list of users younger than 18 years old.

lst Example:

```python

lst = [(1,"hasan", 19), (2, "ali", 16), (6, "david", 25)]
```

````python
lst = [(1,"hasan", 19), (2, "ali", 16), (6, "david", 25)]
def filterFunc(us):
    return us[2]<18
def filterUsers(uslst):
    return list(filter(filterFunc, uslst))

print(filterUsers(lst))
    
````

## Even or Odd
