## Dictionaries

Dictionaries hold store data values in pairs of **keys** and **values**. Dictionaries are declared as:


We can store any data type inside a dict:

````python
player={
    "intKey" : 1,
    "strKey" : "str value",
    "boolKey" : True,
    "listeKey" : [1,2,3,4,5]
}
print(player)
````

We can access elements of a dict using **[]** operator

````python

player["intKey"]
````

````python
player["listeKey"]
````

Elements of a dict are also changable.

````python
player["boolKey"]=False
print(player["boolKey"])
````

We can also add new key-value pairs to dicts.

````python
player["newKey"]=100.0
print(player)
````

Just like in lists and tuples, we can traverse key-value pairs of a dict

````python
for key, value in player.items():
    print(key, "->", value)
````

keys() function gives a list of keys belonging to the dict

````python
for key in player.keys():
    print(key)
````

values() function gives a list of values belonging to the dict

````python
for val in player.values():
    print(val)
````

# Abstract Data Types ( Stack, Queue )

Stacks and queues are lists/arrays of elements. They both support adding and removing elements. Their difference is that;

- Stacks work with **LIFO (Last In First Out)** rule. When we do a remove operation on stack, it will remove the element which we put last
- Queues work with **FIFO (First In First Out)** rule. When we remove an element from queue, it removes the element which we placed earliest.

- In stacks, remove operation is called **pop()** and add operation is **push()**
- In queues, remove operation is called **dequeue()** and add operation is **enqueue()**
- In stacks, we get the element at the top by **top()** function
- In queues, we get the element at the front by **front()** function

## Example Check Parantheses

Define a function named "check_parantheses" which takes a string and returns **True** if none of the parantheses ``(``, brackets ``[`` and braces ``{`` remain open at the end ot the string and they are cloed in correct order and **False** otherwise

sample input output

```python
>>> check_parantheses('There are many online editors for programming languages (including Python).')
True
>>> check_parantheses('(5 + 4) * 7 - [(3 + 2) / 2] % {2 * [-1 / (x - t)]}')
True
>>> check_parantheses('{a, b, [c, d}, f]')
False
>>> check_parantheses('Use a tool (e.g. a spell-checker to spot errors.')
False
```

````python
import stack
def check_parantheses(s):
    st=stack.createStack()
    chropen=['(','[','{']
    chrclose=[')',']','}']

    for c in s:
        if c in chropen:
            stack.push(c, st)
        elif c in chrclose:
            if stack.isEmpty(st):
                return False
            elif chrclose.index(c)!=chropen.index(stack.pop(st)):
                return False
    if not stack.isEmpty(st):
        return False
    return True

check_parantheses('{a, b, [c, d}, f]')
````
